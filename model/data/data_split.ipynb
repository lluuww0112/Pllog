```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("train.csv")
```

```python
data.head()
```

```python
# X: 입력 데이터, y: emotion 라벨
X = data.drop(columns=['emotion'])
y = data['emotion']

# stratify=y 를 사용해야 클래스 비율 동일하게 분배됨
x_train, x_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.1,      # 테스트셋 비율 (원하는 비율로 변경 가능)
    random_state=42,    # 재현성을 위해 고정
    stratify=y          # ★ emotion 기준 층화 분할
)
```

```python
# 다시 하나의 데이터프레임으로 결합하기 (필요한 경우)
train_df = x_train.copy()
train_df['emotion'] = y_train

test_df = x_test.copy()
test_df['emotion'] = y_test
```

```python
print(train_df.shape)
print(test_df.shape)
```

```python
train_df.to_csv("train.csv", index=False)
```

```python
test_df.to_csv("test.csv", index=False)
```
