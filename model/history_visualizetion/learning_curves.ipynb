```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
attn_history = pd.read_csv("./data/Attn_history.csv")
cross_history = pd.read_csv("./data/Cross_history.csv")
```

```python
plt.plot(attn_history["loss"])
plt.plot(cross_history["loss"])
plt.legend(["Attn", "Cross"])
plt.title("Loss")
```

```python
plt.plot(attn_history["recall@10"])
plt.plot(cross_history["recall@10"])
plt.legend(["Attn", "Cross"])
plt.title("Recall_10")
```

```python
plt.plot(attn_history["alignment"])
plt.plot(cross_history["alignment"])
plt.legend(["Attn", "Cross"])
plt.ylim(0, 0.5)
plt.title("Alignment")
```

```python
plt.plot(attn_history["uniformnity"])
plt.plot(cross_history["uniformnity"])
plt.legend(["Attn", "Cross"])
plt.ylim(top=-2.75)
plt.title("Uniformnity")
```

```python
plt.plot(attn_history["val_recall@10"])
plt.plot(cross_history["val_recall@10"])
plt.legend(["Attn", "Cross"])
plt.title("Val_Recall@10")
```
